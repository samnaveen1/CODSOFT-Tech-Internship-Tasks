```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Internship /Credit Card Fraud Detection/creditcard.csv')
data.head()
```

```python
#finding the missing value
data.isnull().sum()
```

```python
# to Split the x and y
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
#to scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA to  dimensions reduction
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
print(f'Original number of features: {X.shape[1]}')
print(f'Reduced number of features: {X_pca.shape[1]}')
```

```python
# Handle class imbalance using SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_pca, y)
```

```python
# Split the data into train and test model
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
from sklearn.ensemble import RandomForestClassifier
```

```python
# Reduce the number of trees
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    max_samples=0.8,
    random_state=42,
    n_jobs=-1
)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# To  Evaluate the model Accuracy
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nAUC-ROC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))
```
