```python
#To load the dataset
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Internship /IBM indian movies/IMDb Movies India.csv',encoding='latin')
data.head()
```

```python
#checking the null values
data.isnull().sum()
#In this dataset, there are a total of 15,000 rows, but around 7,000 to 8,000 rows contain null values. To avoid affecting the model's accuracy, we should exclude columns with a large number of null values.
```

```python
data.shape
```

```python
data.info()
```

```python
# To convert the catagerical variable to numerical
from sklearn.preprocessing import LabelEncoder
obj1=LabelEncoder()
data['Genre']=obj1.fit_transform(data[['Genre']])
data['Director']=obj1.fit_transform(data[['Director']])
data['Actor 1']=obj1.fit_transform(data[['Actor 1']])
data['Year']=obj1.fit_transform(data[['Year']])
data['Rating']=obj1.fit_transform(data[['Rating']])
```

```python
data.head()
```

```python
# To filling the missing values
data['Year']=data['Year'].fillna(data['Year'].mean())
data['Genre']=data['Genre'].fillna(data['Genre'].mean())
data['Director']=data['Director'].fillna(data['Director'].mean())
data['Actor 1']=data['Actor 1'].fillna(data['Actor 1'].mean())
data['Rating']=data['Rating'].fillna(data['Rating'].mean())
```

```python
data.isnull().sum()
```

```python
# to scaling the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for col in ['Year','Genre','Actor 1','Director','Rating']:
  data[[col]] = sc.fit_transform(data[[col]])
```

```python
# to split the x and y and train and test
X=data.iloc[:,[1,3,6,7]]
X
y=data.iloc[:,4]
y
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# To fit the model
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(xtrain,ytrain)
ypred=le.predict(xtest)
```

```python
# To  Evaluate the model
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import cross_val_predict
r2=r2_score(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
```

```python
print("R-squared (R2)Score:",r2)
print("Mean Squared Error(MSE):",mse)
```

Reason :

In this dataset, there are a total of 15,000 rows, but around 7,000 to 8,000 rows contain null values. Since the target variable (y) is the "Rating," the large number of missing values in this column may be the reason for the low accuracy of the model.
